So from the above list we just want to get the eact title weather the movies is action adventures or fantacy, so lets perform some operations.

```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
updated_movies_merge['genres'] = updated_movies_merge['genres'].apply(convert)
```

```python
updated_genres = updated_movies_merge
```

```python
updated_genres.head()
```

lets update the keywords so that we can get some insights of the data

```python
updated_genres['keywords'] = updated_genres['keywords'].apply(convert)
```

```python
updated_keywords = updated_genres
```

```python
updated_keywords.head()
```

lets work on the cast of the movies data. we only want to get the first three lists of the cast to get the info of top three actors of the movies

```python
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
           l.append(i['name'])
           counter+=1
        else:
            break   
    return l
```

```python
updated_keywords['cast'] = updated_keywords['cast'].apply(convert3)
```

```python
updated_cast = updated_keywords
```

```python
updated_cast.head()
```

Now we have to clean the data of the crew column in which we get the insights of the job discription which is director.

```python
def find_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
           l.append(i['name'])
           break
    return l
```

```python
updated_cast['crew'] = updated_cast['crew'].apply(find_director)
```

```python
updated_crew = updated_cast
```

```python
updated_crew.head()
```

lets work on the overview so the we can get the overview of each movie in a separate list

```python
updated_crew['overview'] = updated_crew['overview'].apply(lambda x:x.split())
```

```python
updated_overview = updated_crew
```

```python
updated_overview.head()
```

lets concatinate the names of the few columns because our model gets inturrupted by the two name which beginning is same but its ending is diferent so by concatenating the model will results some actionable insights 

```python
updated_crew['genres'] = updated_crew['genres'].apply(lambda x:[i.replace(" ","") for i in x])
updated_crew['keywords'] = updated_crew['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
updated_crew['cast'] = updated_crew['cast'].apply(lambda x:[i.replace(" ","") for i in x])
updated_crew['crew'] = updated_crew['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
updated_concatenation = updated_crew
```

```python
updated_concatenation.head()
```

lets make a new column of tags of concatenate the whole columns into the one column which is known as tags. for furthur modeling of data

```python
updated_concatenation['tags'] = updated_concatenation['overview'] + updated_concatenation['genres'] + updated_concatenation['keywords'] + updated_concatenation['cast'] + updated_concatenation['crew']
```

```python
tags_column = updated_concatenation
```

```python
tags_column.head()
```

```python
new_df = tags_column[['id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector[0]
```

```python
ps.stem('loving')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Superman')
```
